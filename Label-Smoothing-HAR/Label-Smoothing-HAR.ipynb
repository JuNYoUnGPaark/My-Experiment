```python
import os
import sys
import copy
from pathlib import Path
import random

import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, random_split

from sklearn.metrics import f1_score
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

import warnings
warnings.simplefilter(action="ignore", category=FutureWarning)

# ============================
# 0. 기본 설정 (경로 + device + seed)
# ============================
BASE_DIR = Path.cwd()  # 이 노트북이 있는 폴더가 Label-Smoothing-HAR 라고 가정
DATASET_DIR = BASE_DIR / "HAR_dataset_functions"
sys.path.append(str(DATASET_DIR))

# 너가 만든 dataset 모듈들 import
from uci import UCIHARDataset
from wisdm import WISDMDataset
from pamap2 import PAMAP2Dataset
from mhealth import MHEALTHDataset

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)


def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False


set_seed(42)

# ============================
# 1. 모델 정의 (1D ResNet)
# ============================

class ResidualBlock1D(nn.Module):
    def __init__(self, in_channels, out_channels, kernel_size=3, stride=1):
        super().__init__()
        padding = kernel_size // 2
        self.conv1 = nn.Conv1d(in_channels, out_channels, kernel_size,
                               stride=stride, padding=padding)
        self.bn1 = nn.BatchNorm1d(out_channels)
        self.conv2 = nn.Conv1d(out_channels, out_channels, kernel_size,
                               stride=1, padding=padding)
        self.bn2 = nn.BatchNorm1d(out_channels)

        self.shortcut = nn.Sequential()
        if stride != 1 or in_channels != out_channels:
            self.shortcut = nn.Sequential(
                nn.Conv1d(in_channels, out_channels, kernel_size=1, stride=stride),
                nn.BatchNorm1d(out_channels),
            )

    def forward(self, x):
        out = F.relu(self.bn1(self.conv1(x)))
        out = self.bn2(self.conv2(out))
        out = out + self.shortcut(x)
        out = F.relu(out)
        return out


class ResNet1D(nn.Module):
    def __init__(self, input_dim, num_classes,
                 channels=[64, 128, 128, 256],
                 kernel_size=3):
        super().__init__()

        self.stem = nn.Sequential(
            nn.Conv1d(input_dim, channels[0],
                      kernel_size=kernel_size, padding=kernel_size//2),
            nn.BatchNorm1d(channels[0]),
            nn.ReLU(),
        )

        blocks = []
        in_ch = channels[0]
        for ch in channels:
            blocks.append(ResidualBlock1D(in_ch, ch, kernel_size=kernel_size, stride=1))
            in_ch = ch
        self.blocks = nn.Sequential(*blocks)

        self.gap = nn.AdaptiveAvgPool1d(1)
        self.fc = nn.Linear(channels[-1], num_classes)

    def forward_features(self, x):
        x = self.stem(x)
        x = self.blocks(x)
        x = self.gap(x).squeeze(-1)  # (B, C)
        return x

    def forward(self, x):
        h = self.forward_features(x)
        logits = self.fc(h)
        return logits


# ============================
# 2. 학습 / 평가 루프
# ============================

def train_one_epoch(model, loader, optimizer, criterion_train, device):
    model.train()
    total_loss, total_correct, total_samples = 0.0, 0, 0

    for X, y, _ in loader:
        X = X.to(device)
        y = y.to(device)

        optimizer.zero_grad()
        logits = model(X)
        loss = criterion_train(logits, y)
        loss.backward()
        optimizer.step()

        total_loss += loss.item() * X.size(0)
        preds = logits.argmax(dim=1)
        total_correct += (preds == y).sum().item()
        total_samples += X.size(0)

    return total_loss / total_samples, total_correct / total_samples


@torch.no_grad()
def eval_one_epoch(model, loader, criterion_eval, device):
    """
    eval에서는 label_smoothing 없는 CE를 사용 (순수 CE 기준 Loss / NLL / ECE 계산용)
    """
    model.eval()
    total_loss, total_correct, total_samples = 0.0, 0, 0
    all_logits = []
    all_labels = []
    all_features = []

    for X, y, _ in loader:
        X = X.to(device)
        y = y.to(device)

        # feature + logits 분리 계산 (forward_features + fc)
        features = model.forward_features(X)          # (B, D)
        logits = model.fc(features)                  # (B, C)

        loss = criterion_eval(logits, y)

        total_loss += loss.item() * X.size(0)
        preds = logits.argmax(dim=1)
        total_correct += (preds == y).sum().item()
        total_samples += X.size(0)

        all_logits.append(logits.cpu())
        all_labels.append(y.cpu())
        all_features.append(features.cpu())

    all_logits = torch.cat(all_logits, dim=0)
    all_labels = torch.cat(all_labels, dim=0)
    all_features = torch.cat(all_features, dim=0)

    return total_loss / total_samples, total_correct / total_samples, all_logits, all_labels, all_features


# ============================
# 3. Metric 함수들
# ============================

def compute_nll(logits, labels):
    log_probs = F.log_softmax(logits, dim=1)
    nll = -log_probs[torch.arange(len(labels)), labels].mean().item()
    return nll


def compute_mean_max_confidence(logits):
    probs = F.softmax(logits, dim=1)
    max_conf = probs.max(dim=1).values
    return max_conf.mean().item()


def compute_entropy(logits):
    probs = F.softmax(logits, dim=1)
    ent = -(probs * (probs.clamp(min=1e-12).log())).sum(dim=1)
    return ent.mean().item()


def compute_margin(logits):
    probs = F.softmax(logits, dim=1)
    top2 = torch.topk(probs, k=2, dim=1).values  # (N, 2)
    margin = top2[:, 0] - top2[:, 1]
    return margin.mean().item()


def compute_ece(logits, labels, n_bins=15):
    probs = F.softmax(logits, dim=1)
    confidences, predictions = probs.max(dim=1)
    labels = labels.cpu()
    confidences = confidences.cpu()
    predictions = predictions.cpu()

    ece = 0.0
    bin_boundaries = torch.linspace(0.0, 1.0, n_bins + 1)

    for i in range(n_bins):
        start = bin_boundaries[i]
        end = bin_boundaries[i + 1]
        mask = (confidences > start) & (confidences <= end)
        if mask.sum().item() == 0:
            continue

        conf_bin = confidences[mask].mean().item()
        acc_bin = (predictions[mask] == labels[mask]).float().mean().item()
        ece += (mask.float().mean().item()) * abs(acc_bin - conf_bin)

    return ece


def compute_intra_inter_ratio(features, labels):
    """
    features: (N, D)
    labels: (N,)
    """
    features = features.numpy()
    labels = labels.numpy()
    classes = np.unique(labels)

    centroids = {}
    intra_list = []

    for c in classes:
        mask = labels == c
        feats_c = features[mask]
        mu_c = feats_c.mean(axis=0)
        centroids[c] = mu_c
        dists = np.linalg.norm(feats_c - mu_c, axis=1)
        intra_list.append(dists.mean())

    intra = float(np.mean(intra_list))

    inter_dists = []
    cls_list = list(classes)
    for i in range(len(cls_list)):
        for j in range(i + 1, len(cls_list)):
            c1, c2 = cls_list[i], cls_list[j]
            d = np.linalg.norm(centroids[c1] - centroids[c2])
            inter_dists.append(d)

    inter = float(np.mean(inter_dists)) if len(inter_dists) > 0 else np.nan
    ratio = intra / inter if inter > 0 else np.nan
    return intra, inter, ratio

def compute_f1_macro(logits, labels):
    preds = logits.argmax(dim=1).cpu().numpy()
    y_true = labels.cpu().numpy()
    return f1_score(y_true, preds, average="macro")

# ============================
# 4. 데이터셋 설정 (경로만 수정하면 됨)
# ============================

# !!! 여기 네 로컬 경로로 수정 !!!
uci_path     = r"C:\Users\park9\My_Experiment\HAR_data\UCI-HAR"
wisdm_path   = r"C:\Users\park9\My_Experiment\HAR_data\WISDM_ar_v1.1_raw.txt"
pamap2_path  = r"C:\Users\park9\My_Experiment\HAR_data"
mhealth_path = r"C:\Users\park9\My_Experiment\HAR_data\MHEALTHDATASET"

DATASET_CONFIGS = {
    "uci": {
        "path": uci_path,
        "num_classes": 6,
        "input_dim": 9,
        "type": "uci",
    },
    "wisdm": {
        "path": wisdm_path,
        "num_classes": None,  # activity 종류에 따라 달라짐 → 런타임에서 추출
        "input_dim": 3,
        "type": "wisdm",
    },
    "pamap2": {
        "path": pamap2_path,
        "num_classes": 12,
        "input_dim": 36,  # feature_cols 길이
        "type": "pamap2",
    },
    "mhealth": {
        "path": mhealth_path,
        "num_classes": 12,
        "input_dim": 23,
        "type": "mhealth",
    },
}

# ============================
# 5. 한 데이터셋에 대해 LS 실험 실행 함수
# ============================

def prepare_dataloaders_for_dataset(name, cfg, batch_size=128):
    path = cfg["path"]
    ds_type = cfg["type"]

    if ds_type == "uci":
        full_train = UCIHARDataset(path, split="train")
        test_ds    = UCIHARDataset(path, split="test")

        # train/val split (80/20)
        n_total = len(full_train)
        n_train = int(n_total * 0.8)
        n_val   = n_total - n_train
        train_ds, val_ds = random_split(full_train, [n_train, n_val])

    elif ds_type == "wisdm":
        full = WISDMDataset(path, window_size=80, step_size=40)
        # 클래스 수 추출
        num_classes = int(full.y.max() + 1)
        DATASET_CONFIGS["wisdm"]["num_classes"] = num_classes

        n_total = len(full)
        n_train = int(n_total * 0.8)
        n_val   = int(n_total * 0.1)
        n_test  = n_total - n_train - n_val
        train_ds, val_ds, test_ds = random_split(full, [n_train, n_val, n_test])

    elif ds_type == "pamap2":
        full = PAMAP2Dataset(path, window_size=500, step_size=250)
        n_total = len(full)
        n_train = int(n_total * 0.8)
        n_val   = int(n_total * 0.1)
        n_test  = n_total - n_train - n_val
        train_ds, val_ds, test_ds = random_split(full, [n_train, n_val, n_test])

    elif ds_type == "mhealth":
        full = MHEALTHDataset(path, window_size=128, step_size=64)
        n_total = len(full)
        n_train = int(n_total * 0.8)
        n_val   = int(n_total * 0.1)
        n_test  = n_total - n_train - n_val
        train_ds, val_ds, test_ds = random_split(full, [n_train, n_val, n_test])

    else:
        raise ValueError(f"Unknown dataset type: {ds_type}")

    if ds_type == "uci":
        test_ds = test_ds
    else:
        test_ds = test_ds

    train_loader = DataLoader(train_ds, batch_size=batch_size, shuffle=True,  drop_last=True)
    val_loader   = DataLoader(val_ds,   batch_size=batch_size, shuffle=False, drop_last=False)
    test_loader  = DataLoader(test_ds,  batch_size=batch_size, shuffle=False, drop_last=False)

    return train_loader, val_loader, test_loader


def run_ls_for_dataset(name, cfg, ls_alpha, epochs=10, batch_size=128):
    print(f"\n=== [{name.upper()}] LS α={ls_alpha} ===")

    input_dim = cfg["input_dim"]
    num_classes = cfg["num_classes"]
    if num_classes is None:
        # WISDM에서 runtime에 채워짐
        pass

    train_loader, val_loader, test_loader = prepare_dataloaders_for_dataset(name, cfg, batch_size=batch_size)

    # num_classes가 None이면 여기서 추론
    if cfg["num_classes"] is None:
        # datasets에서 y.max() + 1로 대체 가능하지만,
        # 여기서는 val_loader에서 한 배치 뽑아서 확인
        for _, y, _ in train_loader:
            cfg["num_classes"] = int(y.max().item() + 1)
            num_classes = cfg["num_classes"]
            break

    model = ResNet1D(input_dim=input_dim, num_classes=num_classes).to(device)

    criterion_train = nn.CrossEntropyLoss(label_smoothing=ls_alpha)
    criterion_eval  = nn.CrossEntropyLoss(label_smoothing=0.0)
    optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)

    # 간단히: val 기준으로만 모니터링 (early stop 없이)
    for epoch in range(1, epochs + 1):
        train_loss, train_acc = train_one_epoch(model, train_loader, optimizer, criterion_train, device)
        val_loss, val_acc, _, _, _ = eval_one_epoch(model, val_loader, criterion_eval, device)
        print(f"[Epoch {epoch:02d}] Train Loss={train_loss:.4f}, Acc={train_acc:.4f} | "
              f"Val Loss={val_loss:.4f}, Acc={val_acc:.4f}")

    # 최종 test에서 메트릭 계산
    test_loss, test_acc, test_logits, test_labels, test_features = eval_one_epoch(
        model, test_loader, criterion_eval, device
    )

    nll = compute_nll(test_logits, test_labels)
    ece = compute_ece(test_logits, test_labels, n_bins=15)
    mean_conf = compute_mean_max_confidence(test_logits)
    mean_ent = compute_entropy(test_logits)
    mean_margin = compute_margin(test_logits)
    intra, inter, ratio = compute_intra_inter_ratio(test_features, test_labels)
    f1 = compute_f1_macro(test_logits, test_labels)

    results = {
        "test_acc": test_acc,
        "test_f1": f1,
        "test_nll": nll,
        "test_ece": ece,
        "mean_max_conf": mean_conf,
        "mean_entropy": mean_ent,
        "mean_margin": mean_margin,
        "intra_dist": intra,
        "inter_dist": inter,
        "separability_ratio": ratio,
    }

    return results, test_features, test_labels


# ============================
# 6. 4개 데이터셋 × (LS=0, 0.01) 전체 실행
# ============================
def plot_tsne(features, labels, title="", max_samples=2000):
    if isinstance(features, torch.Tensor):
        features_np = features.numpy()
    else:
        features_np = features
    if isinstance(labels, torch.Tensor):
        labels_np = labels.numpy()
    else:
        labels_np = labels

    N = len(labels_np)
    if N > max_samples:
        idx = np.random.choice(N, max_samples, replace=False)
        features_np = features_np[idx]
        labels_np = labels_np[idx]

    tsne = TSNE(n_components=2, perplexity=30, learning_rate="auto",
                init="random", random_state=42)
    emb = tsne.fit_transform(features_np)

    plt.figure(figsize=(6, 5))
    plt.scatter(emb[:, 0], emb[:, 1], c=labels_np, s=12, alpha=0.7, cmap="tab10")
    plt.title(title)
    plt.xticks([]); plt.yticks([])
    plt.tight_layout()
    plt.show()


ALL_RESULTS = {}
ALL_FEATURES = {}  # t-SNE용 feature 저장

for ds_name in ["wisdm", "pamap2", "uci", "mhealth"]:
    cfg = copy.deepcopy(DATASET_CONFIGS[ds_name])

    # LS 없는 경우
    set_seed(42)
    res_no_ls, feat_no_ls, lab_no_ls = run_ls_for_dataset(ds_name, cfg, ls_alpha=0.0,  epochs=10, batch_size=128)

    # LS 0.01
    set_seed(42)
    res_ls, feat_ls, lab_ls = run_ls_for_dataset(ds_name, cfg, ls_alpha=0.01, epochs=10, batch_size=128)

    ALL_RESULTS[(ds_name, "no_ls")]   = res_no_ls
    ALL_RESULTS[(ds_name, "ls_0.01")] = res_ls

    ALL_FEATURES[(ds_name, "no_ls")]   = (feat_no_ls, lab_no_ls)
    ALL_FEATURES[(ds_name, "ls_0.01")] = (feat_ls, lab_ls)

    # 🔥 바로 t-SNE도 그려보기
    print(f"\n[t-SNE] {ds_name.upper()} / no LS vs LS=0.01")

    f0, l0 = ALL_FEATURES[(ds_name, "no_ls")]
    f1, l1 = ALL_FEATURES[(ds_name, "ls_0.01")]

    plot_tsne(feat_no_ls, lab_no_ls, title=f"{ds_name.upper()} - w/o Label Smoothing")
    plot_tsne(feat_ls,    lab_ls,    title=f"{ds_name.upper()} - + Label Smoothing(0.01)")



# ============================
# 7. 결과 요약 출력
# ============================

print("\n" + "="*80)
print("FINAL SUMMARY (7 metrics per setting)")
print("="*80)

metrics_order = [
    "test_acc",
    "test_f1",
    "test_nll",
    "test_ece",
    "mean_max_conf",
    "mean_entropy",
    "mean_margin",
    "intra_dist",
    "inter_dist",
    "separability_ratio",
]

for ds_name in ["uci", "wisdm", "pamap2", "mhealth"]:
    print(f"\n[{ds_name.upper()}]")
    for setting in ["no_ls", "ls_0.01"]:
        res = ALL_RESULTS[(ds_name, setting)]
        print(f"  - {setting}:")
        for k in metrics_order:
            print(f"      {k:18s} = {res[k]:.4f}")
```
